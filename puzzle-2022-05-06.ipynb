```python
startLetters, letters, coordToLen, coordToWord = init()
```

```python
def show():
    # prints out the letters that we can select from
    # map of row : string for letters
    top = {}
    for idx, column in enumerate(letters):
        word = letters[column]
        spaces = 6 - len(word)  # number of spaces we have to add to the top
        padTo6 = str(idx % 10) + '-' # first two rows should be idx + -
        
        for i in range(spaces):
            padTo6 += ' '

        for letter in word:
            padTo6 += letter + ''
        
        # populate the column appropriately
        for i, letter in enumerate(padTo6):
            if i not in top:
                top[i] = '|'
            top[i] += letter + '|'
            
    for row in top:
        print(top[row])
    
    print('--------------------------------------------------------------------------------')
    
    bottom = {}
    bottom[0] = ''
    bottom[1] = ''
    bottom[2] = ''
    bottom[3] = ''
    bottom[4] = '|▓'
    for coord in coordToLen:
        row, col = coord
        if coord not in coordToWord or coordToWord[coord] == '':
            # We haven't solved this position yet, so fill with spaces
            length = coordToLen[coord]
            for i in range(length):
                bottom[row] += '| '
        else:
            for letter in coordToWord[coord]:
                bottom[row] += '|' + letter
        if len(bottom[row]) < 78:
            bottom[row] += '|▓'
        else:
            bottom[row] += '|'
    
    for row in bottom:
        print(bottom[row])
```

```python
def solve(letterOptions):
    for word in itertools.product(*letterOptions):
        wordStr = ''.join(word)
        if wordStr in dictionary[len(letterOptions)]:
            print(wordStr)
```

```python
def options(coord, with_original=False):
    if coord in coordToLen:
        row, col = coord
        length = coordToLen[coord]
        
        # Figure out which letters we can use
        letterOptions = []
        for i in range(length):
            if with_original:
                letterOptions.append(startLetters[i + col])
            else:
                letterOptions.append(letters[i + col])
        
        print('Possible options:')
        solve(letterOptions)
    else:
        print('options: did not pass a valid coord')
```

```python
def add(word, coord):
    if coord in coordToLen:
        row, col = coord
        length = coordToLen[coord]

        # Figure out which letters we can use
        letterOptions = []
        for i in range(length):
            letterOptions.append(letters[i + col])
            
        # Make sure this word doesn't already exist
        if coord in coordToWord:
            print('add: ' + coordToWord[coord] + ' already exists at coord ' + str(coord) + '\n')
            show()
            return
        
        # Make sure that we're actually allowed to add this word
        for idx, char in enumerate(word):
            if char not in letterOptions[idx]:
                print('add: invalid word! does not have ' + char + ' at ' + str(col + idx))
        
        # Remove the characters from each of the letters
        for idx, char in enumerate(word):
            options = letterOptions[idx]
            replaced = options.replace(char, "", 1)
            letters[idx + col] = replaced

        coordToWord[coord] = word
        print('Added ' + word + ' to coord ' + str(coord) + '\n')
        show()
    else:
        print('add: did not pass a valid coord')
```

```python
def deleteCoord(coord):
    if coord in coordToLen:
        row, col = coord
        length = coordToLen[coord]
            
        # Make sure this word already does exist
        if coord not in coordToWord:
            print('delete: a word does not exist at coord ' + str(coord) + '\n')
            show()
            return
        
        # Figure out which columns we need to add back to
        letterOptions = []
        for i in range(length):
            letterOptions.append(letters[i + col])
        
        # Add the letters back, in sorted order!
        word = coordToWord[coord]
        for idx, char in enumerate(word):
            options = letterOptions[idx]
            added = ''.join(sorted(options + char))
            letters[idx + col] = added


        del coordToWord[coord]
        print('Deleted ' + word + ' from coord ' + str(coord) + '\n')
        show()
    else:
        print('delete: did not pass a valid coord')
```

```python
def delete(word):
    locations = []
    for coord in coordToWord:
        wordAtCoord = coordToWord[coord]
        if wordAtCoord == word:
            locations.append(coord)
    if len(locations) == 0:
        print(word + ' not found in puzzle')
    if len(locations) == 1:
        deleteCoord(locations[0])
    else:
        print('delete not successful: word appears multiple times. try running:')
        for coord in locations:
            print('deleteCoord(' + str(coord) + ')')
    print()
    show()
```

```python
def help():
    print('coord is a (row, column) position in the puzzle. rows')
    print('and columns are 0-indexed.')
    print()
    print('COMMANDS')
    print('------------------------------------------------------')
    print('add(word, coord): adds a word at a specified location.')
    print('  EX: add(\'rank\', (2, 2)) ')
    print()
    print('delete(word): deletes a word from the puzzle')
    print('  If a word appears more than once, you will have to specify which location with deleteCoord.')
    print('  EX: delete(\'rank\')')
    print()
    print('deleteCoord(coord): deletes a word from the puzzle at a specified location.')
    print('  EX: deleteCoord((2, 2))')
    print()
    print('options(coord, with_original=False): displays what words can be created at a coordinate.')
    print('  This does not change the puzzle state.')
    print('  EX: options((2, 2)) shows the possible options for (2, 2) with the current puzzle state.')
    print('  EX: options((2, 2)) shows the possible options for (2, 2) with the original puzzle state.')
    print()
    print('show(): prints the current puzzle board state')
    print()
    print('------------------------------------------------------')
    print('To reset, run the following in a cell:')
    print('startLetters, letters, coordToLen, coordToWord = init()')
```
